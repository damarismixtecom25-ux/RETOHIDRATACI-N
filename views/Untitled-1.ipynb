import pandas as pd
import matplotlib.pyplot as plt

# 1. Creación de los datos 
data = {
    'Nombre': [
        'Jaquelin', 'Xiomara', 'Gael', 'Alvaro', 'Melissa', 'Ieth', 'Angeles', 
        'Jennifer', 'Yocimar', 'Astrid', 'Rodrigo', 'Alam', 'Marco', 'Diego', 
        'Jonathan', 'Jose Luis', 'Joana', 'Kevin', 'Maria', 'Angel Emir', 
        'Danna', 'Magaly', 'Itzel', 'Priscila', 'Nallely'
    ],
    'Edad': [
        17, 17, 16, 19, 17, 17, 18, 17, 18, 17, 17, 17, 17, 17, 17, 17, 17, 
        18, 17, 17, 17, 18, 16, 17, 17
    ]
}

df = pd.DataFrame(data)

# 2. Cálculos de Medidas de Tendencia Central
media = df['Edad'].mean()
mediana = df['Edad'].median()
moda = df['Edad'].mode()[0]

# 3. Cálculos de Medidas de Dispersión
rango = df['Edad'].max() - df['Edad'].min()
varianza = df['Edad'].var(ddof=0) # Varianza poblacional como en el Excel
desviacion_std = df['Edad'].std(ddof=0)

# 4. Tabla de Frecuencias
frecuencias = df['Edad'].value_counts().sort_index().reset_index()
frecuencias.columns = ['Edad', 'fi']
frecuencias['hi'] = frecuencias['fi'] / len(df)
frecuencias['Fi'] = frecuencias['fi'].cumsum()
frecuencias['Hi'] = frecuencias['hi'].cumsum()

# --- Resultados ---
print("--- MEDIDAS ---")
print(f"Media: {media:.2f}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")

print("\n--- DISPERSIÓN ---")
print(f"Rango: {rango}")
print(f"Varianza: {varianza:.4f}")
print(f"Desviación Estándar: {desviacion_std:.10f}")

print("\n--- TABLA DE FRECUENCIAS ---")
print(frecuencias)

# 5. Generación de Gráfica
plt.figure(figsize=(8, 5))
bars = plt.bar(frecuencias['Edad'].astype(str), frecuencias['fi'], color='#b4a7d6', edgecolor='grey')

# Añadir etiquetas de datos sobre las barras
plt.bar_label(bars)

plt.title('Frecuencia de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
